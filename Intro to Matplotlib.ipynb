```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot([1,5,3,7])
plt.show()
```

```python
plt.plot([1,5,7,3],[2,6,4,8])
plt.title('Test PLot',fontsize=8,color='g')
plt.xlabel('number n')
plt.ylabel('number o')
plt.show()
```

```python
plt.figure(figsize=(2,5)) #in inches
plt.plot([1,5,7,3],[2,6,4,8],'ro')
plt.axis([0,20,0,20])  # xmin,xmax,ymin,ymax
plt.annotate('square it',(3,8))
plt.show()
```

```python
plt.clf()  # clear figure
x=np.arange(4)
y=[8.8,5.2,3.6,5.9]
plt.xticks(x,('Ankit','Hans','Joe','Flaco'))
plt.bar(x,y,color='r')
plt.show()
```

```python
plt.bar(x,y,color=['Lime','r','k','tan'])
plt.show()
```

```python
d={'Red O':np.random.rand(10),
   'Grn X':np.random.rand(10)}
df=pd.DataFrame(d)
df.plot(style=['ro','gx'])
plt.show()
```

```python
N=50
x=np.random.rand(N)
y=np.random.rand(N)
colors=np.random.rand(N)
sizes=(30*np.random.rand(N))**2
plt.scatter(x,y,s=sizes,c=colors,alpha=0.5)
plt.show()
```

```python
fig=plt.figure()
fig.suptitle('Subplots')
fig.add_subplot(221)
plt.plot([np.log(n) for n in range(1,10)])
fig.add_subplot(222,facecolor='y')
fig.add_subplot(223)
fig.add_subplot(224)
plt.show()
```

```python
fig,plots=plt.subplots(2,sharex=True)
fig.suptitle('Sharing Axis')
x=range(0,200,5)
y=[n**0.8 for n in x]
plots[0].plot(x,y,color='r')
plots[1].scatter(x,y)
plt.show()
```

```python
plt.figure(figsize=(4,3),dpi=100)
plt.plot([245,170,148,239,161,196,112,258])
plt.axis([0,7,0,300])
plt.title('Flight Data')
plt.xlabel('speed')
plt.savefig('flights.png')
plt.show()
```
